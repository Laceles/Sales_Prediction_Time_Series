```python
str_cols=['Promo2SinceWeek', 'Promo2SinceYear', 'PromoInterval',
            'CompetitionOpenSinceYear', 'CompetitionOpenSinceMonth']
for i in str_cols:
    store_df[i].fillna(0,inplace=True)
```

```python
sns.heatmap(store_df.isnull());
```

```python
store_df.hist(bins=30,figsize=(20,20), color='b');
```

```python
all_dfs=pd.merge(open_train_df,store_df,how='inner', on='Store')
all_dfs.head()
```

```python
correlacoes=all_dfs.corr()
f,ax=plt.subplots(figsize=(20,20))
sns.heatmap(correlacoes, annot=True);
```

```python
correlacoes=correlacoes['Sales'].sort_values()
correlacoes
```

```python
all_dfs['Year']=pd.DatetimeIndex(all_dfs['Date']).year
all_dfs['Month']=pd.DatetimeIndex(all_dfs['Date']).month
all_dfs['Day']=pd.DatetimeIndex(all_dfs['Date']).day
```

```python
all_dfs.head()
```

```python
axis=all_dfs.groupby('Month')[['Sales']].mean().plot(figsize=(10,5),marker='o',color='r')
axis.set_title('Média de vendas por mês')
```

```python
axis=all_dfs.groupby('Month')[['Customers']].mean().plot(figsize=(10,5),marker='^',color='b')
axis.set_title('Média de clientes por mês')
```

```python
axis=all_dfs.groupby('Day')[['Sales']].mean().plot(figsize=(10,5),marker='o',color='r')
axis.set_title('Média de vendas por dia')
```

```python
axis=all_dfs.groupby('Day')[['Customers']].mean().plot(figsize=(10,5),marker='o',color='b')
axis.set_title('Média de clientes por dia')
```

```python
axis=all_dfs.groupby('DayOfWeek')[['Sales']].mean().plot(figsize=(10,5),marker='o',color='r')
axis.set_title('Média de vendas por dia da semana')
```

```python
axis=all_dfs.groupby('Day')[['Customers']].mean().plot(figsize=(10,5),marker='o',color='b')
axis.set_title('Média de vendas por dia da semana')
```

```python
fig,ax=plt.subplots(figsize=(20,10))
all_dfs.groupby(['Date','StoreType']).mean()['Sales'].unstack().plot(ax=ax)
```
